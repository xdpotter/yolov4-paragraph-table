```python
import os

# Define base path to your dataset
base_path = '/content/drive/MyDrive/yolov4/invoice/'

# Paths to image directories
train_images_path = base_path + 'images/train/'
valid_images_path = base_path + 'images/valid/'

# List image files
train_images = [os.path.join(train_images_path, img) for img in os.listdir(train_images_path) if img.endswith('.jpg')]
valid_images = [os.path.join(valid_images_path, img) for img in os.listdir(valid_images_path) if img.endswith('.jpg')]

# Create train.txt
with open(base_path + 'train.txt', 'w') as f:
    f.write("\n".join(train_images) + "\n")

# Create valid.txt
with open(base_path + 'valid.txt', 'w') as f:
    f.write("\n".join(valid_images) + "\n")
```
